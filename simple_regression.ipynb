<h1> Simple Linear Regression model </h1>

<h2> Importing libraries </h2>

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import sklearn
```

<h2> Importing the dataset </h2>

```python
data = pd.read_csv('./archive/simple.csv')
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
data
```

<h2> Dividing dataset into the Training set and the test set </h2>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

<h2> Traning the Simple Linear Regression model on the Training set </h2>

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
%matplotlib inline
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, model.predict(x_train), color = "blue")
plt.title = "SAT VS GPA ( Training set )"
plt.xlabel('SAT')
plt.ylabel('GPA')
```

```python
%matplotlib inline
plt.scatter(x_test, y_test, color = "red")
plt.plot(x_train, model.predict(x_train), color = "blue")
plt.title = "SAT VS GPA ( Test set )"
plt.xlabel('SAT')
plt.ylabel('GPA')
```
